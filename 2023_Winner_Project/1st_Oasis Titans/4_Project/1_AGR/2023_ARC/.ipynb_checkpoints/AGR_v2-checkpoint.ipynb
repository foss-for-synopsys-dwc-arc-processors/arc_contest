## Load and preprocess training an testing dataset

```python
class Config():
    # Data path
    IMAGE_DIR = "./train_img"
    LABEL_DIR = "./train_label/label.csv"
    BG_DIR = "./gauge_img/gauge.png"
    
    # Dataset parameters
    SPLIT = 0.2
    K_SPLIT = 5
    
    # Batch size
    BATCH_SIZE = 32 #8
    
    # Image size
    IMG_HEIGHT = 100
    IMG_WIDTH = 100
    
    # Model parameters
    NUM_CLASSES = 72
    IN_CHANNEL = 1
    INPUT_SHAPE = (IMG_HEIGHT, IMG_WIDTH, IN_CHANNEL)
    
    # Learning parameters
    lr = 1e-3
    num_epochs = 50
    loss_func = "categorical_crossentropy"
```

```python
img_gauge = img.imread(Config.BG_DIR)


img_gauge = Image.open(Config.BG_DIR).convert('RGBA')
img_gauge = img_gauge.resize((Config.IMG_HEIGHT, Config.IMG_WIDTH), Image.BILINEAR)
# plt.imshow(img_gauge)

bg = img_gauge.convert('L')
bg = np.array(bg)
plt.imshow(bg,cmap='gray')
```

```python
def load_data(img_path, lab_path, split=0.2):
    img_files = glob.glob(img_path + "/*")
    lab_df = pd.read_csv(lab_path)['degree'].to_numpy()
    data_x = []
    data_y = []
    
    # concatenate
    for idx, img_file in enumerate(img_files):
        img = Image.open(img_file)
        img = img.convert('L')
        img = np.array(img, dtype=int)
#         img = img / 255.0 # Normalize
        
        lab = np.array(lab_df[idx],dtype=int)
        lab = np.expand_dims(lab,axis=0)
        data_x.append(img)
        data_y.append(lab)
    
    # Train & val split
    split_size = int(split * len(img_files))
    train_x, test_x = train_test_split(data_x, test_size=split_size, random_state=42)
    train_y, test_y = train_test_split(data_y, test_size=split_size, random_state=42)
    
    train_x = np.array(train_x)
    train_y = np.array(train_y)
    
    test_x = np.array(test_x)
    test_y = np.array(test_y)

    return train_x, train_y, test_x, test_y
```

```python
# Loading the dataset
train_x, train_y, test_x, test_y = load_data(Config.IMAGE_DIR, Config.LABEL_DIR)
```

```python
fig = plt.figure(figsize=(10,7))
for i in range(15):  
    ax = fig.add_subplot(3, 5, i+1)
    img = Image.fromarray(np.uint8(train_x[i])) #
    img = img.convert('RGBA')
    
    ax.imshow(img, cmap=plt.get_cmap('gray'))
    ax.set_title('class: {y}'.format(y=np.squeeze(train_y[i])))#
    plt.axis('off')
```

## Load and process training and testing dataset

```python
# Dataset preprocessing #1

# Transfer to nparray
train_y = to_categorical(train_y, Config.NUM_CLASSES)
test_y = to_categorical(test_y, Config.NUM_CLASSES)
```

## Model define and create

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(Config.IMG_HEIGHT,Config.IMG_WIDTH,1)))
model.add(BatchNormalization())

model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
# model.add(Dropout(0.2))


model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
# model.add(Dropout(0.2))


# model.add(Conv2D(512, (3, 3), padding='same', activation='relu'))
# model.add(BatchNormalization())
# model.add(MaxPooling2D(pool_size=(2, 2)))
# model.add(Dropout(0.2))


model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(Config.NUM_CLASSES, activation='sigmoid'))  
```

```python
# Training model

# Define optimizer loss function and merics 
# opt = SGD(lr=Config.lr, momentum=0.9)
opt = 'adam'

model.compile(optimizer=opt,
              loss='categorical_crossentropy',
              metrics=['accuracy'])


train_x = train_x.reshape(-1, Config.IMG_HEIGHT, Config.IMG_WIDTH, 1) #

# Set training
history = model.fit(train_x, train_y,
          validation_split = 0.2,
          batch_size = Config.BATCH_SIZE,
          verbose = 1,
          epochs = Config.num_epochs
          )
```

```python
# accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
